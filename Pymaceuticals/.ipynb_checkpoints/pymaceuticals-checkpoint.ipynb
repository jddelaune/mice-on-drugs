### Analysis

Our study measured the results of the treatment of mice with ten different drugs.  There were 249 mice in the study; one had to be excluded due to duplicate data which was clearly erroneously recorded, leaving 248 mice.

Mice were observed at multiple time points throughout their treatment. The minimum number of total observations per drug was 148 and the most promising two drugs, Capomulin and Ramicane, had over 200 observations each. (see Figure: Drug Regimen, cell 12)

The two most effective drugs in the study by far were Capomulin and Ramicane. They saw extremely similar performance, with a mean tumor volume at the end of the treatment period of 40.68 and 49.22 respectively. (See summary statistics in cell 9 and boxplot in cell 20) The performance of both drugs was far better than any of the other drugs included in the study.

The graph shown in cell 23 illustrates the reduction in tumor volume over the entire Capomulin treatment regimen for a single example mouse.

The graph in cell 51 shows that for mice treated with Capomulin, the average tumor volume is strongly correlated with the mouse's weight. Unfortunately, someone would probably notice if we tried to re-run the study and give the skinny mice Capomulin and give the fat mice Ramicane.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_metadata_df = pd.DataFrame(mouse_metadata)
study_results_df = pd.DataFrame(study_results)
frames = [mouse_metadata_df, study_results_df]
mice_drugs_df = study_results_df.merge(mouse_metadata_df, on='Mouse ID')

# Display the data table for preview
mice_drugs_df
```

```python
# Checking the number of mice.
mice_count = mice_drugs_df['Mouse ID'].nunique()
mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse_data_s = mice_drugs_df.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)
mice_drugs_df[duplicate_mouse_data_s]
```

```python
# Get the mouse IDs that have duplicate info
duplicate_mice = mice_drugs_df[duplicate_mouse_data_s]['Mouse ID'].unique()
duplicate_mice
```

```python
# Optional: Get all the data for any duplicate mouse ID. 
all_dupe_mouse_data_s = mice_drugs_df['Mouse ID'].isin(duplicate_mice)
all_dupe_mouse_data_s

duplicate_mice_rows_df = mice_drugs_df[all_dupe_mouse_data_s]
duplicate_mice_rows_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mice_drugs_deduped_df = mice_drugs_df[~all_dupe_mouse_data_s]
mice_drugs_deduped_df
```

```python
# Checking the number of mice in the clean DataFrame.
mice_drugs_deduped_df['Mouse ID'].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

mean_tumor_vol_s = mice_drugs_deduped_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
median_tumor_vol_s = mice_drugs_deduped_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
tumor_vol_var_s = mice_drugs_deduped_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
tumor_vol_stdev_s = mice_drugs_deduped_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
tumor_vol_stderr_s = mice_drugs_deduped_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()

tumor_vol_dict = {'Mean Tumor Volume' : mean_tumor_vol_s, 'Median Tumor Volume' : median_tumor_vol_s, 
                 'Tumor Volume Variance' : tumor_vol_var_s, 'Tumor Volume Std. Dev.' : tumor_vol_stdev_s,
                 'Tumor Volume Std. Err.' : tumor_vol_stderr_s}

tumor_vol_df = pd.DataFrame(tumor_vol_dict)
tumor_vol_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
mice_drugs_deduped_df.groupby('Drug Regimen')["Tumor Volume (mm3)"].agg(['mean', 'median', 'var', 'std', 'sem'])
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Start by getting the last (greatest) timepoint for each mouse
maxtime_df = mice_drugs_deduped_df.groupby(['Mouse ID']).max()['Timepoint'].reset_index()


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_timepoint_df = maxtime_df.merge(mice_drugs_deduped_df, how='inner', on=['Mouse ID', 'Timepoint'])
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Capomulin, Ramicane, Infubinol, and Ceftamin
promising_drugs = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list (of lists!) to fill with tumor vol data (for plotting)
promising_drug_tumor_vol = []

# Loop through each promising drug
for drug in promising_drugs:

    # Locate the rows which contain mice on this drug and get the tumor volumes as a series
    mice_on_this_drug_s = final_timepoint_df['Drug Regimen'] == drug
    tumor_vols_for_this_drug = final_timepoint_df[mice_on_this_drug_s]['Tumor Volume (mm3)']

    # add this subset to the list for plotting
    promising_drug_tumor_vol.append(tumor_vols_for_this_drug)
    # Calculate the IQR for this drug  
    q1 = tumor_vols_for_this_drug.quantile(q=0.25)
    q3 = tumor_vols_for_this_drug.quantile(q=0.75)
    iqr = q3 - q1
    
    # Determine outliers for this drug using upper and lower bounds and print them
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    
    print("The lower bound for " + drug + " is " + str(lower_bound) + ".")
    print("The upper bound for " + drug + " is " + str(upper_bound) + ".")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(promising_drug_tumor_vol, labels=promising_drugs)
plt.show()
```

## Line and Scatter Plots

```python
# Find unique Mouse IDs for mice on Capomulin
cap_mice_df = mice_drugs_deduped_df.loc[mice_drugs_deduped_df['Drug Regimen'] == 'Capomulin', ["Mouse ID"]]
pd.unique(cap_mice_df["Mouse ID"])
```

```python
mouse_l509_df = mice_drugs_deduped_df.loc[mice_drugs_deduped_df["Mouse ID"] == "l509"]
mouse_l509_df
```

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
# x-axis is the timepoint
x_axis = mouse_l509_df["Timepoint"]
# y-axis is the tumor volume
y_axis = mouse_l509_df["Tumor Volume (mm3)"]
plt.plot(x_axis, y_axis)

plt.title("Mouse l509, treated with Capomulin")
plt.xlabel("Timepoint")
plt.ylabel("Tumor volume (mm3)")
plt.show()
```

```python
# get list of unique mouse IDs that were treated with Capomulin -- filter the dataframe instead
cap_mouseIDs = pd.unique(cap_mice_df["Mouse ID"])
cap_mouseIDs
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
# Create empty lists to hold data
avg_cap_mouse_weight = []
avg_cap_tumor_vol = []

# Loop through each mouse
for mouse in cap_mouseIDs:
    #Locate rows with this mouse and add weight & tumor volume to lists
    current_mouse_df = mice_drugs_deduped_df.loc[mice_drugs_deduped_df["Mouse ID"] == mouse]
    avg_weight = current_mouse_df["Weight (g)"].mean()
    avg_tumor_vol = current_mouse_df["Tumor Volume (mm3)"].mean()
    
    avg_cap_mouse_weight.append(avg_weight)
    avg_cap_tumor_vol.append(avg_tumor_vol)
    
print(avg_cap_mouse_weight)
print(avg_cap_tumor_vol)
```

```python
#Create scatter plot
plt.scatter(avg_cap_mouse_weight, avg_cap_tumor_vol)

plt.ylim(34,46)
plt.xlim(14,26)

plt.title("Average Tumor Volume and Weight for Mice on Capomulin")
plt.xlabel("Average Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
corr = st.pearsonr(avg_cap_mouse_weight, avg_cap_tumor_vol)[0]
print("The correlation between mouse weight and the average tumor volume is " + str(corr))
slope, intercept, r, p, stderr = st.linregress(avg_cap_mouse_weight, avg_cap_tumor_vol)
```

```python
cap_s = mice_drugs_deduped_df['Drug Regimen'] == 'Capomulin'
cap_df = mice_drugs_deduped_df[cap_s]
cap_weight_s = cap_df.groupby('Mouse ID').mean(numeric_only=True)['Weight (g)']

cap_y = cap_weight_s.values * slope + intercept
cap_y

```

```python
plt.plot(cap_weight_s.values, cap_y, c="red")

plt.scatter(avg_cap_mouse_weight, avg_cap_tumor_vol)
```
